```python
import pandas as pd
import numpy as np
```

```python
# Create a sample DataFrame with a timestamp column
data = {'timestamp': ['2025-10-20 08:30:00', '2025-10-20 09:15:00',
                    '2025-10-20 10:45:00', '2025-10-21 11:00:00',
                    '2025-10-21 12:30:00', '2025-10-21 13:00:00',
                    '2025-10-21 14:00:00', '2025-10-22 09:00:00',
                    '2025-10-22 09:05:00'],
        'flow_id': [1, 2, 3, 4, 5, 6, 7, 8, 9]}

dx = pd.DataFrame(data)

# Convert the 'timestamp' column to datetime objects
dx['timestamp'] = pd.to_datetime(dx['timestamp'])
```

```python
dx
```

```python
flows_per_day = dx.groupby(dx['timestamp'].dt.date).size()
```
